```python
#Взять датасет из домашнего задания №2. Проверить гипотезу о том, 
#в каком варианте теста (control/personalization) больше конверсия (converted) 
#и значимо ли это отличие статистически
import pandas as pd
import scipy.stats as stats
from scipy.optimize import Bounds
from scipy.optimize import minimize
import numpy as np

df_1 = pd.read_csv('marketing_campaign.csv')
df_2 = pd.read_csv('subscribers.csv')
df_3 = pd.read_csv('users.csv')
df_12 = df_1.merge(df_2, left_on = ['user_id'],
                    right_on =['user_id'], how = 'left')
df = df_12.merge(df_3, left_on = ['user_id'],
                    right_on =['user_id'], how = 'left')
df.head()
```

```python
dfp = pd.pivot_table(df, index = "variant", columns = "converted", values = "user_id", aggfunc = "count")
dfp.head()
```

H0: разницы в конверсии нет

H1: (вариант B) дает большую конверсию

```python
result = stats.chi2_contingency(dfp, correction = False)
chisq, pvalue = result[:2]

print(f"{chisq=}, {pvalue=}")
```

```python
#Цех может производить стулья и столы. На производство стула идет 5 единиц материала, на производство стола - 20 
#единиц (футов красного дерева). Стул требует 10 человеко-часов, стол - 15. 
#Имеется 400 единиц материала и 450 человеко-часов. Прибыль при производстве стула - 45 долларов США, 
#при производстве стола - 80 долларов США. Сколько надо сделать стульев и столов, чтобы получить 
#максимальную прибыль?

def value(x):
    return - 45*x[0] - 80*x[1]
```

```python
neq_cons = {'type': 'ineq',
            'fun': lambda x: np.array ([400 - 5  * x [0] - 20 * x [1], 
                                        450 - 10  * x [0] - 15 * x [1]])
}
```

```python
bnds = Bounds([0, 0], [np.inf, np.inf])
```

```python
x0 = np.array([10, 10])
res = minimize(value, x0, method='SLSQP', constraints=neq_cons, bounds=bnds)
print(res.x)
```
